# Segregating Discrete and continuos Feature

# Discrete

```python
#we are assuming 32 as threshold were feature having less than 32 is discrete
dist_fea = [fea for fea in num_col if len(df[fea].unique())<32]
dist_fea
```

```python
for feature in dist_fea:
    sns.countplot(data=df,x=feature,hue='Classes')
    plt.show()
```

# continous

```python
cont_fea = [fea for fea in num_col if fea not in dist_fea]
cont_fea
```

```python
import scipy.stats as stats
```

```python
for i in cont_fea:
    plt.figure(figsize=(15,6))
    plt.subplot(121)
    sns.histplot(data=df,x=i,kde=True,bins=30,color='green')
    plt.title("{}'s distribution".format(i),fontweight='bold')
    
    plt.subplot(122)
    stats.probplot(df[i],dist='norm',plot=plt)
    plt.title("{}'s Q-Q plot".format(i),fontweight='bold')
    
    
```

# Comparing Continous Numerical feature with region

```python
plt.figure(figsize=(20,40))
for i in enumerate(cont_fea):
    plt.subplot(4,2,i[0]+1)
    sns.set(rc={'figure.figsize':(7,7)})
    sns.histplot(data=df,x=i[1],kde=True,bins=30,color='blue',hue='Region')
    plt.title("{} vs Region".format(i[1],fontweight='bold'))
```

```python
plt.figure(figsize=(20,40))
for i in enumerate(cont_fea):
    plt.subplot(4,2,i[0]+1)
    sns.set(rc={'figure.figsize':(7,7)})
    sns.histplot(data=df,x=i[1],kde=True,bins=30,color='blue',hue='Classes')
    plt.title("{} Vs Classes".format(i[1],fontweight='bold'))
```
